Alguma coluna tem mais de 60% de dados nulos? Assim podemos desconsiderar estas colunas para a nossa anÃ¡lise.

```python
max_rows = df["RespondentID"].count()
print("% Missing Data: ")
print(((1 - df.count() / max_rows) * 100) >= 60)
```

```python
#excluir as colunas com 60% ou mais de dados faltantes

df.drop(["NEWJobHunt", "NEWJobHuntResearch", "NEWLearn"],
axis=1,
inplace=True)
```

```python
print((1 - df.count() / max_rows) * 100)
```

```python
from matplotlib.pyplot import bar


df[["RespondentID", "Country"]].groupby("Country").count()

#vamos analisar duas das colunas com missing data, mas nÃ£o tÃ£o significante assim (prÃ³x. de 10%)
missing_data = df[["Employment", "DevType"]].isnull().groupby(df["Country"]).sum().reset_index()

A = sns.catplot(data=missing_data,
kind="bar",
x="Country", y="Employment",
aspect=2,
height=6
)

B = sns.catplot(data=missing_data,
kind="bar",
x="Country", y="DevType",
aspect=2,
height=6
)
```

Os dados nÃ£o parecem ser tÃ£o discrepantes assim.
MCAR = missing complete at random
Neste caso podemos utilizar Pairwise e manter os dados das demais colunas intactas.

```python
df.dropna(subset=["Employment", "DevType"],
inplace=True,
how="any")
```
